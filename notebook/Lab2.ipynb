Script for second task

```python
%%bash
cd lab2

cat > seqMeasurer
#!/bin/bash

> lengths.txt

for file in *.fa; do
  grep ">" "$file" >> lengths.txt

  grep -v ">" "$file" | tr -d /n | wc -c >> lengths.txt

  echo >> lengths.txt

done
```

```python
%%bash
cd lab2
chmod u+x seqMeasurer
./seqMeasurer
cat lengths.txt
```
